```python
import dudo

utils = dudo.train(5000000, print_nodes=False)
```

```python
import plotly.express as px
import numpy as np

running_avg = np.cumsum(utils) / np.arange(1, len(utils) + 1)
# drop 999 out of 1000 points
running_avg_sam = running_avg[::1000]

px.scatter(y=running_avg_sam, title='Util rate over time').show()
```
